```python
import pandas as pd

# Load the dataset
df = pd.read_csv('train.csv')

# Calculate the correlation coefficients
correlation_matrix = df.corr()

# Get the correlation coefficients with the last column
correlation_with_last_column = correlation_matrix.iloc[:, -1]

# Get the absolute values of the correlation coefficients
absolute_correlation = correlation_with_last_column.abs()

# Sort the columns by absolute value of correlation coefficients
sorted_columns = absolute_correlation.sort_values(ascending=False).index

feature_used = 15
# Print the sorted column names
features = ""
for feature in sorted_columns[1:feature_used+1]:
    features += feature + ", "
print(features)
```
